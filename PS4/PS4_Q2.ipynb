```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
xVals = yVals = []
nSteps = 10000
xMin = 0
xMax = 1
fMin = 0
fMax = 2
```

```python
def fn(x):
    return 2*x/(xMax**2)
```

```python
for i in range(nSteps):
    x = random.random()
    xVals.append(x)
    yVals.append(fn(x))
```

```python
plt.hist(yVals, bins=100, histtype="step")
plt.title("Sawtooth PDF")
plt.xlabel("f(x) values")
```

```python
acceptX = []

for i in range(10000):
    x = xMin + random.random()*(xMax - xMin)
    u = random.random()*fMax
    if u < fn(x):
        acceptX.append(x)
        
plt.hist(acceptX, bins=100)
plt.title("Sawtooth PDF with accepted x values")
plt.xlabel("x values")
```
